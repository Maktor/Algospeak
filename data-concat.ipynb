```python
import pandas as pd

hatebase = pd.read_csv("C:\\Users\\marlo\\OneDrive\\Desktop\\Coding\\hatebase.csv")
bluesky = pd.read_csv("C:\\Users\\marlo\\OneDrive\\Desktop\\Coding\\bluesky.csv")
```

```python

hatebase_new = (hatebase.loc[:, ["tweet", "data"]].rename(columns={"tweet": "text"}))
hatebase_new = (hatebase_new.loc[:, ["text"]].assign(data="hatebase"))

bluesky_new = (bluesky.loc[:, ["text"]].assign(data="bluesky"))

```

```python

combined = pd.concat([hatebase_new, bluesky_new], ignore_index=True)


combined.to_csv("combined.csv", index=False)



```
